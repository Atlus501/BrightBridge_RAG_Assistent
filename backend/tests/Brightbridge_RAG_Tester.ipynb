This is a notebook that tests the api endpoints for the fastapi version of the project. It sends requests to the port forwarded by ngrok to test the api endpoints and whether they give appropriate responses

```python
import requests

endpoints = ["/respond", "/store_past_context", "/get_past_context"]
payloads = [
    {
        "past_conv": [],
        "prompt" : "What are some good relationship therapy options?"
    },

    {
        "session_id" : None,
        "password": "soidfjskfdkwortiwurtxkvj",
        "actor_id": "test-user",
        "conv_to_be_saved": [],
    }
]
```

```python
import requests
import dotenv
import os

url = os.getenv("NGROK_URL")

"""
Function to test the server endpoint.
Params: api_endpoint (str) -- server api endpoint
        payload (dict) -- request payload
Returns: None
"""
def test_server_endpoint(api_endpoint, payload):
  try:
    r = requests.post(f"{url}{api_endpoint}", json=payload)
    return r
  except Exception as e:
    print(e)
    return {"error" : str(e)}
```

```python
r = test_server_endpoint(endpoints[0], payloads[0])
print(r)
```
